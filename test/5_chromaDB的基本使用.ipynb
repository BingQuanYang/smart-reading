```python
from langchain_community.embeddings import DashScopeEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
chroma_db = Chroma(
    persist_directory="./chroma_db",
    #集合  表
    collection_name="test-collection",
    #嵌入模型
    embedding_function=DashScopeEmbeddings(),
    collection_metadata={"hnsw:space":"cosine"}
)
```

```python
data = [
    Document(page_content="猫坐在垫子上"),
    Document(page_content="一只猫坐在垫子上休息"),
    Document(page_content="今天天气很好")
]
chroma_db.add_documents(
    documents=data,
    ids=[f"id-{i}" for i in range(1, len(data) + 1)]
)
"""
id  原声文本 向量
id-1  猫坐在垫子上   embedding_function.embed_query(猫坐在垫子上)
"""
```

```python
#相识度搜索
user_input = "猫在哪?"
# result = chroma_db.similarity_search(user_input, 2)
result = chroma_db.similarity_search_with_relevance_scores(user_input, 3)
print(type(result))
print(result)
```
